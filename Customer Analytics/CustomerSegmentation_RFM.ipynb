```python
for col in ['Quantity', 'Price', 'TotalPrice']:
    
    Q1 = df[col].quantile(0.01)
    Q3 = df[col].quantile(0.99)
    IQR = Q3-Q1
    upper = Q3 + 1.5*IQR
    lower = Q1 - 1.5*IQR
    
    if df[(df[col] > upper)|(df[col] < lower)].any(axis=None) :
        print(col, 'yes')
        print(df[(df[col] > upper) | (df[col] < lower)].shape[0])
    else :
        print(col, 'no')
```

## Customer Segmentation with RFM Scores

> - Recency : time since customer last purchase (today's date - last purchase)
> - Frequency : total number of purchases
> - Monetary(Monetary Value) : total spending by the customer

```python
df.head()
```

```python
df.info()
```

```python
# to fix customer ID
df['Customer ID'] = df['Customer ID'].astype(int)
```

#### DATE Variables

```python
print(df['InvoiceDate'].min())
print(df['InvoiceDate'].max())
```

```python
# set the today's date as the maximum day of the dataset
import datetime as dt
todaydate = dt.datetime(2010,12,9)
todaydate
```

#### the last shopping dates of each customer

```python
df.groupby('Customer ID').agg({'InvoiceDate':'max'}).head() # the last day > max
```

### customer segmentation with RFM

#### Recency

```python
(todaydate - df.groupby('Customer ID').agg({'InvoiceDate' : 'max'})).head()
```

```python
temp_df = (todaydate - df.groupby('Customer ID').agg({'InvoiceDate' : 'max'}))
temp_df.rename(columns={'InvoiceDate' : 'Recency'}, inplace = True)
temp_df.head()
```

```python
recency_df = temp_df['Recency'].apply(lambda x : x.days)
recency_df.head()
```

#### Frequency

```python
# Customer ID 별 Invoice COUNT
temp_df = df.groupby(['Customer ID', 'Invoice']).agg({'Invoice':'count'})
temp_df.head()
```

```python
# Customer ID 별 Invoice COUNT SUM 총 주문 횟수 
temp_df.groupby('Customer ID').agg({'Invoice' : 'sum'})
```

```python
freq_df = temp_df.groupby('Customer ID').agg({'Invoice' : 'sum'})
freq_df.rename(columns = {'Invoice': 'Frequency'}, inplace=True)
freq_df.head()
```

#### Monetary

```python
monetary_df = df.groupby('Customer ID').agg({'TotalPrice':'sum'})
monetary_df.rename(columns={'TotalPrice':'Monetary'}, inplace=True)
monetary_df.head()
```

```python
print(recency_df.shape, 
     freq_df.shape,
     monetary_df.shape)
```

#### join tables

```python
rfm = pd.concat([recency_df, freq_df, monetary_df], axis=1)
rfm.head()
```

### Scoring according to the most Recent, the Cyclic, and the monetary expenditure

```python
rfm['RecencyScore'] = pd.qcut(rfm['Recency'], 5, labels = [5,4,3,2,1])
rfm['FrequencyScore'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels = [1,2,3,4,5])
rfm['MonetaryScore'] = pd.qcut(rfm['Monetary'], 5, labels = [1,2,3,4,5])
rfm.head()
```

```python
rfm.info()
```

```python
rfm['RFM_SCORE'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str) + rfm['MonetaryScore'].astype(str)
rfm.head()
```

```python
rfm.describe().T
```

### regex segmentation 
- considering r and f

```python
seg_map = {
    r'[1-2][1-2]' : 'Hibernating',
    r'[1-2][3-4]' : 'At Risk',
    r'[1-2]5' : 'Can\'t Loose',
    r'3[1-2]' : 'About to Sleep',
    r'33' : 'Need Attention',
    r'[3-4][4-5]' : 'Loyal Customers',
    r'41' : 'Promising',
    r'51' : 'New Customers',
    r'[4-5][2-3]' : 'Potential Loyalists',
    r'5[4-5]' : 'Champions'
}
```

- Adding Segment Variable

```python
rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
rfm.head()
```

```python
rfm[['Segment', 'Recency', 'Frequency','Monetary']].groupby('Segment').agg(['mean','count'])
```
